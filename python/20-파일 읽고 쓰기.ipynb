```python
# 파일 입출력
# 웹 크롤링 -> 데이터 클렌징 -> 유용한 데이터셋 도출 -> 데이터 입력 및 인공지능 학습

# 파일 경로 복사
# /content/sample_data/california_housing_test.csv

# r - read 읽기, w - write (덮어)쓰기, a - add 추가
```

```python
# 파일 생성하기
f = open("newfile.txt", 'w')
f.close()
```

```python
f = open("newfile.txt", 'w')
message = ''

for i in range(10):
    message += "Hello, my friends.\n"
f.write(message)
f.close()
```

```python
f = open("newfile.txt", 'w')
f.write('Hello, my friends.')
f.close()
```

```python
# 위 코드를 더 간단하게 작성하는 방법:
with open('newfile.txt', 'w') as f:
    f.write('Fighting~!')
```

```python
# 저장된 파일 내용 읽어오기

f = open("newfile.txt", 'r')
message = f.read()
print(message)
f.close()
```

```python
f = open("newfile.txt", 'r')
line = f.readline()
print(line)
print(type(line))
```

```python
while line:
    line = f.readline()      # 한줄씩 출력됨
    print(line)
f.close()
```

```python
# list 값 입력
list = [10, 20, 30]
with open('newfile.txt', 'w') as f:
    f.write(str(list))
```

```python
# csv 파일
list = ['10', '20', '30']
with open('newfile.csv', 'w') as f:
    f.write(','.join(list))
```
