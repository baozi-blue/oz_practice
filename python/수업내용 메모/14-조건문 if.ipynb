```python
# if문이란? -> 특정 조건에 따라 코드를 실행하는 문법
# 문법과 주의사항 :
# 1) 들여쓰기(indentation) - 들여쓰기는 언제나 같은 깊이로 해야 한다.
#                        -> 공백문자 4개 (권장), spacebar/tab 혼용 X
# 2) 콜롬 ':' - 빠지면 안된다.


num = int(input("빵 몇개 사실 건가요?"))

if num > 3:
    print("개당 1800원입니다.")
else:
    print("개당 2000원입니다.")
```

```python
# if 중첩문
num2 = 10
if num2 > 5:
    if num2 < 15:
        print("해당 숫자는 5보다 크고 15보다 작습니다.")
```

```python
# 여러개 조건식 설정하기
# 연산자:
# and, or, not, in, not in
# <, >, ==, !=, >=, <=
# 조건문에 아무일도 하지 않게 설정하고 싶을때 -> pass 사용

a, b = 10, 30

if a==10 and b==30:
    print("ok")
else:
    print("not ok")
```

```python
# 다양한 조건을 판단하는 elif
# elif의 개수는 제한없음

a, b = 20, 40
if a==10:
    pass
elif b==30:
    pass
else:
    print("not ok")
```

```python
# 조건부 표현식
score = 61
message = "ok" if score>= 60 else "not ok"
print(message)
```
