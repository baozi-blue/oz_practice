```python
# 키워드 매개변수
# kwargs는 ‘keyword arguments’의 약자이며 args와 마찬가지로 관례적으로 사용한다.

def print_kwargs(**kwargs):
    print(kwargs)


print_kwargs(kwargs_1="Shark", kwargs_2=4.5, kwargs_3=True)
```

```python
def print_values(**kwargs):
    for key, value in kwargs.items():
        print("The value of {} is {}".format(key, value))

print_values(kwargs_1="Shark", kwargs_2=4.5, kwargs_3=True)
```

```python
# 주의사항: 함수의 리턴값은 언제나 하나이다.
# 함수는 return문을 만나는 순간 결괏값을 돌려준 다음 함수를 빠져나가게 된다.

def add_and_mul(a,b):
    return a+b, a*b

result = add_and_mul(3,4)
print(result)
```

```python
# 결과값을 분리하여 저장하고 싶을때
result1, result2 = add_and_mul(3,4)
print(result1)
print(result2)
```

```python
def oz(x, y, h):
    print(x*y*h)

oz(10, 20, 10)
```

```python
# x, y, h 의 초기값을 미리 지정할 수 있고, 지정한 값을 추후 변경할 수 있다.
# 초기화하고 싶은 매개변수는 항상 뒤쪽에 놓아야 한다. 안그러면 오류 난다.

def oz(x, y=10, h=20):
    print(x*y*h)

oz(10) # x = 10
oz(10, y=20) # y값 수정
oz(x = 5, y = 5, h = 5) # x, y, h값 모두 수정
```
