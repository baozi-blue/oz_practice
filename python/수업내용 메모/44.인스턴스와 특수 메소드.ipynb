```python
# isinstance() : 첫번째 매개변수에는 인스턴스(객체), 두번째 매개변수에는 클래스를 입력해주면
# 두번째 매개변수로 입력한 클래스 기반으로 인스턴스가 생성되었는지 True, False로 반환해준다.

class Student:
    def __init__(self):
        pass

student = Student()

print(isinstance(student, Student))
print(isinstance(student, int))
```

```python
# isinstance() 함수를 이용해 여러개의 클래스에서 생성된 인스턴스를 구분해 처리하는 코드

class Python:
    def python(self):
        print("파이썬 수강중입니다.")

class Java:
    def java(self):
        print("자바 수강중입니다.")

programming_subjects = [Python(), Python(), Java()]

for subject in programming_subjects:
    if isinstance(subject, Python):   # True
        subject.python()
    elif isinstance(subject, Java):    # True
        subject.java()

```
