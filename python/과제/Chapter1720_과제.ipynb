과제_1: BungeoppangShop 클래스 만들기

- __init__: 초기 재고, 가격, 판매량을 설정합니다.
- check_stock: 현재 재고를 출력합니다.
- process_order: 주문을 처리하고 재고를 업데이트합니다.
- admin_mode: 관리자 모드에서 재고를 추가합니다.
- calculate_total_sales: 총 매출을 계산하고 출력합니다.

```python
bread_price = {"팥붕어빵": 1000, "슈크림붕어빵": 1200, "초코붕어빵": 1500}
stock = {"팥붕어빵": 10, "슈크림붕어빵": 8, "초코붕어빵": 5}
total_order = []
total_amount = 0

class BungeoppangShop:
    def __init__(self, bread_type, bread_count):
        global total_order, total_amount
        total_order.append((bread_type, bread_count))  
    def check_stock(self, bread_type, bread_count):
        if bread_type in stock.keys():
            bread_count = int(bread_count)
            print('구매 가능합니다.')
            print(f'팥붕어빵 재고 : {stock["팥붕어빵"]}개, 슈크림붕어빵 재고 : {stock["슈크림붕어빵"]}개, 초코붕어빵 재고 : {stock["초코붕어빵"]}개')
        else:
            print("등록되지 않은 제품입니다.")
    def process_order(self, bread_type, bread_count):
        if stock[bread_type] >= bread_count:
            stock[bread_type] -= bread_count
        else:
            print("재고 부족합니다.")
        if stock[bread_type] == 0:
            print("재고 다 떨어졌습니다. 판매를 중지 합니다.")
    def admin_mode(self, bread_type, add_num):
        if bread_type in stock.keys():
            add_num = int(add_num)
            stock[bread_type] += add_num
            print(f'팥붕어빵 재고 : {stock["팥붕어빵"]}개, 슈크림붕어빵 재고 : {stock["슈크림붕어빵"]}개, 초코붕어빵 재고 : {stock["초코붕어빵"]}개')
        else:
            print("등록되지 않은 제품입니다.")
    def calculate_total_sales(self, bread_type, bread_count):
        global total_order, total_amount
        total_order.append((bread_type, bread_count)) 
        total_amount += bread_count*bread_price[bread_type]
        print(f'제품명 : {bread_type}, 주문수량 : {bread_count}개, 주문금액 : {bread_count*bread_price[bread_type]}원')    
        
```

```python
order1 = BungeoppangShop("팥붕어빵", 2)
add1 = BungeoppangShop("슈크림붕어빵", 10)

order1.check_stock("팥붕어빵", 2)
order1.process_order("팥붕어빵", 2)
order1.calculate_total_sales("팥붕어빵", 2)
add1.admin_mode("슈크림붕어빵", 10)
```
