```python
# 함수 사용 이유 :
# 반복되는 부분이 있을 경우, ‘반복적으로 사용되는 가치 있는 부분’을 한 뭉치로 묶어
# ‘어떤 입력값을 주었을 때 어떤 결괏값을 리턴해 준다’라는 식의 함수로 작성하는 것이다.


# 함수의 구조

def a(i):
    a = i + 1
    return a

a(7)
```

```python
# 매개변수와 인수의 차이점 :
# 매개변수(parameter)는 함수에 입력으로 전달된 값을 받는 변수,
# 인수(arguements)는 함수를 호출할 때 전달하는 입력값을 의미한다.

def add(a, b):  # a, b는 매개변수
    return a+b   # return 뒤 문구 실행되지 않음, return 없는 함수의 결과값은 none으로 출력됨

print(add(3, 4))  # 3, 4는 인수
```

```python
def count(a, b, c):
    amount = a*1000 + b*1200 + c*1500
    return amount

print(f'총 {count(1,1,1)}원입니다.')
```

```python
# 지역변수와 번역변수  --> 따로 공부 필요!!
```

```python
def harvest():
    vege = {'A level' : 0, 'B level' : 0, 'C level' : 0}

    def onion():
        vege['A level'] += 1
    def spinach():
        vege['A level'] += 1
        vege['B level'] += 2
    def tomato():
        vege['A level'] += 2
        vege['B level'] += 3
        vege['C level'] += 1

    onion()
    spinach()
    tomato()

    return vege

print(harvest())
```
