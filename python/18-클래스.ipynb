```python
# 함수 여러개 설정하기

result1 = 0
result2 = 0

def add1(num):  # 계산기1
    global result1
    result1 += num
    return result1

def add2(num):  # 계산기2
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))

```

```python
# class로 함수 여러개 동시에 설정하기

class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

cal1 = Calculator()  # 인스턴스
cal2 = Calculator()  # 인스턴스

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

```python
class greeting:
    def hello(self):
        print("hello")

    def hi(self):
        print("hi")

obj = greeting()

obj.hello()
obj.hi()
```

```python
class person:
    def __init__(self, name, age, address):
        self.msg = 'Hello'
        self.name = name
        self.age = age
        self.address = address

    def hello(self):
        print(f'{self.msg} I am {self.name}.')

student = person('Maria', 20, 'Seoul')
student.hello()

print('Name:', student.name)
print('Age:', student.age)
print('Address:', student.address)
```

```python
# 클래스의 상속
# 기본 클래스 (base class) -> 파생 클래스 (derived class)
# 중복된 메소드는 overwrite된다.  마지막 메소드만 인정.


class Car(object):
    max_speed = 300
    max_people = 5
    def move(self, x):
        print("달리고 있습니다.")
    def stop(self):
        print("멈췄습니다.")

class HybridCar(Car):
    battery = 1000
    battery_unit = "kWh"

k5 = HybridCar()
k5.max_speed       # max_speed 값을 상속 받음.

print(dir(k5))     # 상속 받은 기능 모두 확인 가능
```
