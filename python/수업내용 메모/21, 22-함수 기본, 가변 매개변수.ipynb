```python
# \n 값을 이용하여 줄바꿈 작동 가능
# 반복되는 횟수 및 문구를 지정받아 출력하는 함수 :

def  oz(message, n):
    for i in range(n):
        print(message)
```

```python
# 여러개 입력값을 받는 함수 만들기

def oz (n, *values):
    for i in range(n):
        for value in values:
            print(value)
        print()  # 보기 편하기 한줄 띄워주기

oz(3, "오늘도", "화이팅","내일도", "화이팅")
```

```python
def add_many(*args):
    result = 0
    for i in args:
        result = result + i # args에 입력 받은 모든 값을 더한다.
    return result

result = add_many(1, 3, 5, 7, 9)
print(result)
```

```python
 # 더하기 곱하기 골라서 실행하는 함수

 def add_mul(choice, *args):
     if choice == "add":   # 매개변수 choice에 "add"를 입력받았을 때
         result = 0
         for i in args:
             result = result + i
     elif choice == "mul":   # 매개변수 choice에 "mul"을 입력받았을 때
         result = 1
         for i in args:
             result = result * i
     return result

result1 = add_mul('add', 1,2,3,4,5)
print(result1)
result2 = add_mul('mul', 1,2,3,4,5)
print(result2)
```
