```python
# range 생성
def oz(start, end, term):
    for i in range(start, end, term):
        print(i, " ")   # 숫자 사이 한칸씩 띄어쓰기 및 줄바뀜

oz(1, 10, 3)
oz(start=5, end=100, term=10)
```

```python
def oz(start, end, term):
    for i in range(start, end, term):
        print(i, " ", end="")   # 숫자 사이 한칸씩 띄어쓰기 및 줄바뀜 없음
    print()   # 출력 결과 줄바뀜

oz(1, 10, 3)
oz(start=5, end=100, term=10)
```

```python
# 재귀 함수를 이용해서 팩토리얼 구현
# 팩토리얼 ! : 5! = 5*4*3*2*1 = 120

def factorial(n):
    output = 1
    for i in range(1, n+1):
        output += i
    return output

n = int(input("구하고자 하는 팩토리얼의 수를 입력해주세요"))
print(f'{n}의 팩토리얼 결과는 {factorial(n)}입니다.')
```

```python
# 피보나치 수열 구하기
# n값이 10보다 커지면 계산 count수도 기하급수적으로 늘어난다

count = 0

def fibonacci(n):
    print(f'피보나치 수열 {n}을 구하는 중입니다.')
    global count
    count += 1

    if n == 1:
        return 1
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n = int(input("구하고자 하는 피보나치 수열의 길이를 입력해주세요"))
fibonacci(n)
print(f'피보나치 수열 {n}의 길이는 {count}입니다.')
```
