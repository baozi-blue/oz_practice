```python
class CreateOzStudent:

    count = 0

    def __init__(self, name, python, database, Django, AWS):
        self.name = name
        self.python = python
        self.database = database
        self.Django = Django
        self.AWS = AWS

        CreateOzStudent.count += 1
        print(f'{CreateOzStudent.count}번째 수강생의 정보가 추가되었습니다.')

    def get_sum(self):
        return self.python + self.database + self.Django + self.AWS

    def get_avg(self):
        return self.get_sum() / 4

    def get_string(self):
        return f'{self.name}\t{self.get_sum()}\t{self.get_avg()}'


# 데이터 입력 : CreateOzStudent() 생성자를 이용해 리스트 안에 데이터 생성
oz_students = [
    CreateOzStudent('Jason', 4, 3, 3, 2),
    CreateOzStudent('Kim', 4, 5, 2, 4),
    CreateOzStudent('Mathew', 3, 4, 4, 1),
    CreateOzStudent('Jane', 2, 3, 1, 5)
]

print('이름', '총점', '평균', sep="\t")

for student in oz_students:
    print(student.get_string())
```

```python
class CreateOzStudent:

    students = []   # 추가

    @classmethod   # 추가
    def print(cls):
        print('이름\t총점\t평균')
        for student in cls.students:
            print(str(student))

    def __init__(self, name, python, database, Django, AWS):
        self.name = name
        self.python = python
        self.database = database
        self.Django = Django
        self.AWS = AWS
        CreateOzStudent.students.append(self)  # 추가

    def get_sum(self):
        return self.python + self.database + self.Django + self.AWS

    def get_avg(self):
        return self.get_sum() / 4

    def __str__(self):
        return f'{self.name}\t{self.get_sum()}\t{self.get_avg()}'


# 데이터 입력 (선언 부분 삭제)
CreateOzStudent('Jason', 4, 3, 3, 2)
CreateOzStudent('Kim', 4, 5, 2, 4)
CreateOzStudent('Mathew', 3, 4, 4, 1)
CreateOzStudent('Jane', 2, 3, 1, 5)


# 출력
CreateOzStudent.print()    # 추가
```
