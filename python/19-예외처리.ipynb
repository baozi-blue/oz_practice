```python
# 예외 코드 작성 "except"

y = [10, 20, 30]
try:
    index, x = map(int, input('나눌 대상 및 숫자를 입력하세요: ').split(','))
    print(y[index]/x)
except ZeroDivisionError:
    print('숫자를 0으로 나눌 수 없습니다.')
except IndexError as e:
    print('잘못된 인덱스입니다.', e)
else:
    print("성공하셨습니다")         # else 로도 결과값 출력 가능 (순서만 다름)
finally:
    print("코드 실행이 끝났습니다.")    # 무조건 실행되는 코드
```
