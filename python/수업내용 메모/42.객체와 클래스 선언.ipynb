```python
oz_students = [
    {'name': 'Jason', 'python': 4, 'database': 3, 'Django': 3, 'AWS': 2},
    {'name': 'James', 'python': 4, 'database': 5, 'Django': 2, 'AWS': 4},
    {'name': 'Justin', 'python': 3, 'database': 4, 'Django': 4, 'AWS': 1},
    {'name': 'Matthew', 'python': 2, 'database': 3, 'Django': 1, 'AWS': 5},
]

print('이름', '총점', '평균', sep="\t")    # sep="\t" 한칸씩 띄어주기

for student in oz_students:
    ability_sum = student['python'] + student['database'] + student['Django'] + student['AWS']
    ability_avg = ability_sum / 4
    print(student['name'], ability_sum, ability_avg, sep="\t")
```

```python
# 여러가지 속성(이름, 과목, 점수 등등)을 가지고 있는 대상을 객체라고 함
# 프로그래밍 언어에서 좀더 넓은 의미로 속성을 가질 수 있는 모든 것을 객체라고 함
```

```python
def create_oz_students(name, python, database, Django, AWS):
    oz_students = {
        'name': name,
        'python': python,
        'database': database,
        'Django': Django,
        'AWS': AWS
    }
    return oz_students
```

```python
# 데이터 입력

create_oz_students('Jason', 4, 3, 3, 2)
create_oz_students('James', 4, 5, 2, 4)
create_oz_students('Justin', 3, 4, 4, 1)
create_oz_students('Matthew', 2, 3, 1, 5)
```

```python
def get_sum(student):
    return student['python'] + student['database'] + student['Django'] + student['AWS']

def get_avg(student):
    return get_sum(student) / 4

def to_string(student):
    return f'{student["name"]}\t{get_sum(student)}\t{get_avg(student)}'
```

```python
print('이름', '총점', '평균', sep="\t")

for student in oz_students:
    print(to_string(student))
```
