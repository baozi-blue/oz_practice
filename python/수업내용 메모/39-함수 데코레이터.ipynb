```python
# 함수 데코레이터는 함수 앞뒤에 붙일 내용이나 반복할 내용이 있을 때 사용된다.

def trace(func):                             # 호출할 함수를 매개변수로 받음
    def wrapper():
        print('함수 시작')
        func()                               # 매개변수로 받은 함수를 호출
        print(func.__name__, '함수 끝')       # __name__으로 함수 이름 출력
    return wrapper                           # wrapper 함수 반환

@trace    # @데코레이터
def hello():
    print('hello')

@trace    # @데코레이터
def world():
    print('world')

hello()    # 함수를 그대로 호출
world()    # 함수를 그대로 호출
```
